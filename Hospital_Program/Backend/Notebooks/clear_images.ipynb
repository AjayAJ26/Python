```python
import cv2
from PIL import Image
```

```python
img=cv2.imread(r"docs\Screenshot 2026-05-30 132758.png",flags=cv2.IMREAD_GRAYSCALE)
print(img)
```

```python
_,new_img=cv2.threshold(img,150,255,cv2.THRESH_BINARY)
Image.fromarray(new_img).show()
```

```python
img2=cv2.imread(r"docs\Screenshot 2026-05-30 134430.png",flags=cv2.IMREAD_GRAYSCALE)
print(img2)
```

```python
_,new_img2=cv2.threshold(img2,150,255,cv2.THRESH_BINARY)
Image.fromarray(new_img2).show()
```

```python
new_img2=cv2.adaptiveThreshold(
    img2, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    67,
    19
)
Image.fromarray(new_img2).show()
```
