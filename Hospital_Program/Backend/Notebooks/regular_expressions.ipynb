```python
import re
```

```python
text="pathion phone number 7304657930 and bill is 130$ "
pattern = '\d{10}'

match =re.findall(pattern,text)
match
```

```python
text = "Dr John Smith, M.D 2 Non-Important Street, New York,  7304657930 or Phone (000)-111-2222"
pattern= '\(\d{3}\)-\d{3}-\d{4}|\d{10}'
match = re.findall(pattern,text)
match
```

```python
text = "pathion phone number 7304657930 and bill is 130$"
pattern= '(\d{10})\D+(\d+)\$'
match = re.search(pattern,text)
match
```

```python
phone_number,bill_amount=match.groups()
```
